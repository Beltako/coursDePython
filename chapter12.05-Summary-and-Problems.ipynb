 $y = r(1 - \cos{\phi})$
 
 where $\phi$ is the number of radians that the wheel has rolled through.

# Problèmes

1. Une cycloïde est la courbe tracée par un point situé sur le bord d'une roue roulant sur une surface plane. Les coordonnées $(x,y)$ d'une cycloïde générée à partir d'une roue de rayon $r$ peuvent être décrites par les équations paramétriques :

 $x = r(\phi - \sin{\phi})$

 $y = r(1 - \cos{\phi})$
 
 où $\phi$ est le nombre de radians parcourus par la roue.
 
 Générez un tracé de la cycloïde pour $0 \le \phi \le 2\pi$ en utilisant 1000 incréments et $r = 3$. Donnez à votre intrigue un titre et des étiquettes. Activez la grille et modifiez les limites des axes pour rendre le tracé soigné.
 
2. Considérez la fonction suivante :

 $y(x) = \sqrt{\frac{100(1-0.01x^2)^2 + 0.02x^2}{(1-x^2)^2 + 0.1x^2}}.$

Utilisez une discrétisation suffisamment fine en *x* pour que le tracé apparaisse lisse. Donnez à chaque axe de tracé des étiquettes et un titre. Allumez la grille. Quelle intrigue semble transmettre le plus d’informations ?.
