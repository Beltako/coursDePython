# 3.6 Résumé


1. Une fonction est un ensemble autonome d’instructions conçues pour effectuer une tâche spécifique.
2. Une fonction possède son propre bloc mémoire pour ses variables. Des informations peuvent être ajoutées au bloc mémoire d'une fonction uniquement via les variables d'entrée d'une fonction. Les informations peuvent quitter le bloc mémoire de la fonction uniquement via les variables de sortie d'une fonction.
3. Une fonction peut être définie au sein d’une autre fonction, appelée fonction imbriquée. Cette fonction imbriquée n'est accessible que par la fonction parent.
4. Vous pouvez définir une fonction anonyme en utilisant le mot-clé lambda, ce qu'on appelle la fonction lambda.
5. Vous pouvez attribuer des fonctions à des variables à l'aide de poignées de fonction.

